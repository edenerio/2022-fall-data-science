## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
condition = df['price'] < 100

df['price is less than 100'] = np.where(condition, 1,0)
statement = df['price is less than 100'] == 1

#for cleaner result, only count specific column
df[statement]['id'].count()
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.
condition_2=df['neighbourhood_group'] == 'Brooklyn'
df_bk = pd.DataFrame(df[condition_2])

#for cleaner result, only count specific column
df_bk['id'].count()
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
statement_2 = df_bk['price is less than 100'] == 1

#for cleaner result, only count specific column
df_bk[statement_2]['id'].count()
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
list_of_names=["Michael","David","John","Daniel"]

statement_3=df['host_name'].isin(list_of_names)

df[statement_3]

```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 
def adjusted_price(x):
    return x+100

statement_4=df['neighbourhood']=='Williamsburg'
df["adjusted_price"] = df['price']
df.loc[statement_4,'adjusted_price']=df["price"].apply(adjusted_price)

df[statement_4].head()
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  
#4 kinds of room types: Entire home/apt, Hotel room, Private room, Shared room
statement_5=df['room_type']=='Private room'
statement_6=df['room_type']=='Shared room'
total_listing=df['id'].count()

print_val_private=(df[statement_5]['id'].count() / total_listing)*100
print_val_shared=(df[statement_6]['id'].count() / total_listing)*100

print(f"Percentage of Private rooms are {print_val_private}%, and shared is {print_val_shared}%")
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
gb=df.groupby('neighbourhood_group')
gb['id'].count()

```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
gb['price'].mean()

```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

#gb already done at this point
gb['price'].agg(['min','max'])

```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
gb_room_type=df.groupby(['neighbourhood_group', 'room_type'])
gb_room_type['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
gb['price'].agg(['count','min','max','mean','median','std'])


```

# Use the grammys.csv data for the next section of questions.

1. Who was won Album of the Year in 2016?


2. Who won Best Rap Album in 2009?


3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?

```python
#  1. Who was won Album of the Year in 2016?
df_grammys=pd.read_csv('./data/grammys.csv')

winners=df_grammys['winner']==1
df_winners=df_grammys[winners]
look_year=df_winners['year']==2016
look_category=df_winners['category']=='Album of the Year'
df_winners[look_year & look_category]['workers']
```

```python
# 2. Who won Best Rap Album in 2009?
look_year_rap=df_winners['year']==2009
look_category_rap=df_winners['category']=='Best Rap Album'
df_winners[look_year_rap & look_category_rap]['workers']
```

```python
# 3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?
nominated=df_grammys['workers'].str.contains('Kendrick Lamar', na=False)

nominations=df_grammys[nominated]
wins=nominations['winner']==True

nums_of_nominations=nominations['year'].count()
nums_of_wins=nominations[wins]['year'].count()

print(f"Kendrick Lamar was nominated {nums_of_nominations} times and won {nums_of_wins} of them")
```
